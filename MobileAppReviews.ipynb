#2. Cleaning App version

```python
df['App Version'].unique()
```

```python
filtered_df = df[~df['App Version'].str.contains("build", na=False)]
filtered_df
```

```python
def clean_app_version(version):
  """Splits on 'build' if present, takes the first part, then trims whitespace."""
  version = str(version)  # Convert to string
  if "build" in version:
    version = version.split("build")[0]  # Split on 'build' and take the first part
  version = version.strip()  # Trim whitespace
  return version

df['App Version'] = df['App Version'].apply(clean_app_version)
df['App Version'].unique()
```

```python
filtered_df = df[df['App Version'] != '1']
filtered_df
```

Check if there are more sub versions to the sub versions and eliminating them

```python
def has_two_dots(version):
  """Checks if a string contains exactly two dots."""
  return version.count('.') == 2

# Filter out rows where 'App Version' is '1'
filtered_df = df[df['App Version'] != '1']

# Check if all remaining rows have two dots
all_have_two_dots = filtered_df['App Version'].apply(has_two_dots).all()

if all_have_two_dots:
  print("All rows (except those with 'App Version' = '1') have two dots in 'App Version'")
else:
  print("Some rows (other than those with 'App Version' = '1') do not have two dots in 'App Version'")
```

```python
def split_at_second_dot(version):
    """Splits the string at the second dot and takes the first part, then trims."""
    first_dot_index = version.find('.')
    if first_dot_index != -1:
        second_dot_index = version.find('.', first_dot_index + 1)
        if second_dot_index != -1:
            version = version[:second_dot_index].strip()
    return version

df['App Version'] = df['App Version'].apply(split_at_second_dot)
df['App Version'].unique()
```
